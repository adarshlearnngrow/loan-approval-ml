## Lets start by dealing with missing data

### Categorial Data

### Missing NAME_TYPE_SUITE is filled with Unspecified

```python
app_data['NAME_TYPE_SUITE'] = app_data['NAME_TYPE_SUITE'].fillna("Unspecified")
```

```python
app_data['NAME_TYPE_SUITE'].unique()
```

```python
app_data[app_data['OWN_CAR_AGE'].isnull()]['FLAG_OWN_CAR'].value_counts()
```

### Imputation Plan for OWN_CAR_AGE
- All owners who do not own a car have missing values in the OWN_CAR_AGE column.
- There is one record where the person owns a car, but the car's age is missing.

```python
app_data.loc[app_data['FLAG_OWN_CAR'] == False, 'OWN_CAR_AGE'] = 0
```

```python
app_data.loc[(app_data['FLAG_OWN_CAR'] == True) & 
             (app_data['OWN_CAR_AGE'].isnull()), 'OWN_CAR_AGE'] = app_data[app_data['FLAG_OWN_CAR'] == True]['OWN_CAR_AGE'].mode()[0]
```

### Occuputation

```python
app_data[app_data['OCCUPATION_TYPE'].isnull()]['TARGET'].value_counts()
```

Most of the missing occuputation type application were rejected, so it may be an important parameter in considering loan application. So where occuputaion is not mentioned we will impute Unknown

```python
app_data['OCCUPATION_TYPE'] = app_data['OCCUPATION_TYPE'].fillna('Unknown')
```

### Family Member Count

```python
app_data['CNT_FAM_MEMBERS'] = app_data['CNT_FAM_MEMBERS'].fillna(0)
```

### Dropping residential information about client

```python
app_data.drop(app_data.columns[44:91], axis=1, inplace=True)
```

### Circle observation, for the same record all the 4 columns are empty, so imputing 0

```python
col = ['OBS_30_CNT_SOCIAL_CIRCLE', 'DEF_30_CNT_SOCIAL_CIRCLE', 'OBS_60_CNT_SOCIAL_CIRCLE', 'DEF_60_CNT_SOCIAL_CIRCLE']
app_data[col] = app_data[col].fillna(0)
```

### Python read bit data type as boolean, so converting it into int

```python
app_data[app_data.columns[49:69]] = app_data[app_data.columns[49:69]].astype(int)
app_data[['TARGET', 'FLAG_OWN_CAR', 'FLAG_OWN_REALTY']] = app_data[['TARGET', 'FLAG_OWN_CAR', 'FLAG_OWN_REALTY']].astype(int)
app_data[app_data.columns[22:28]] = app_data[app_data.columns[22:28]].astype(int)
app_data[app_data.columns[34:40]] = app_data[app_data.columns[34:40]].astype(int)
```

### Using -1 to represent missing values for the AMT columns,  especially since it clearly indicates that no data is available for those records.

```python
app_data[app_data.columns[-6:]] = app_data[app_data.columns[-6:]].fillna(-1)
```

### While loading data into sql, it didnt read 0 values, and imputed 0

```python
app_data['DAYS_REGISTRATION'] = app_data['DAYS_REGISTRATION'].fillna(0)
app_data['DAYS_ID_PUBLISH'] = app_data['DAYS_ID_PUBLISH'].fillna(0)
```

### 56% of EXT_SOURCE_1 is missing, imputing will contaminate the data, so deleteing that column 

```python
app_data.drop('EXT_SOURCE_1', axis=1, inplace=True)
```

### Dealing with EXT_SOURCE_2 and EXT_SOURCE_3 missing values

```python
app_data[['EXT_SOURCE_2', 'EXT_SOURCE_3']].columns
```

```python
fig, axes = plt.subplots(2, 2, figsize = (10,8))

for i,j in enumerate(['EXT_SOURCE_2', 'EXT_SOURCE_3']):
    sns.histplot(data=app_data[app_data['TARGET'] == 0], x=j, ax=axes[i, 0], kde=True, color='green', alpha=0.3)
    axes[i, 0].axvline(app_data[app_data['TARGET'] == 0][j].mean(), ls='--', color='red')
    axes[i, 0].axvline(app_data[app_data['TARGET'] == 0][j].median(), ls='--')
    
    sns.histplot(data=app_data[app_data['TARGET'] == 1], x=j, ax=axes[i, 1], kde=True, color='green', alpha=0.3)
    axes[i, 1].axvline(app_data[app_data['TARGET'] == 1][j].mean(), ls='--', color='red')
    axes[i, 1].axvline(app_data[app_data['TARGET'] == 1][j].median(), ls='--')   
                         
    fig.suptitle("Distribution of External Source", fontsize=16)
    
plt.tight_layout()  
```

### The data is not normally distributed, so will impute median in  missing places

```python
app_data['EXT_SOURCE_2'].median()
```

```python
app_data['EXT_SOURCE_2'] = app_data['EXT_SOURCE_2'].fillna(app_data['EXT_SOURCE_2'].median())
app_data['EXT_SOURCE_3'] = app_data['EXT_SOURCE_3'].fillna(app_data['EXT_SOURCE_3'].median())
```

```python
app_data['DAYS_LAST_PHONE_CHANGE'] = app_data['DAYS_LAST_PHONE_CHANGE'].fillna(0)
```

```python
app_data['AMT_GOODS_PRICE'] = app_data['AMT_GOODS_PRICE'].fillna(app_data['AMT_GOODS_PRICE'].dropna().median())
app_data['AMT_ANNUITY'] = app_data['AMT_ANNUITY'].fillna(app_data['AMT_ANNUITY'].dropna().median())
```
