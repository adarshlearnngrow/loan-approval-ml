```python
from imports import *
from sklearn.model_selection  import train_test_split 
```

```python
query = "select * from cleaned_application_ml"
app_data = sql_connect(query)
app_data.head()
```

```python
X = app_data.drop(columns=['TARGET'])
y = app_data['TARGET']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
train_data = pd.concat([y_train, X_train], axis=1)
test_data = pd.concat([y_test, X_test], axis=1)
```

```python
engine = write_sql_conn_create()
train_data.to_sql('train_data', engine, if_exists='replace', index=False)
```

```python
test_data.to_sql('test_data', engine, if_exists='replace', index=False)
```
